```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('penguins_raw.csv')
```

```python
df = df.drop(['studyName', 'Sample Number','Region','Island','Stage','Individual ID','Clutch Completion', 'Date Egg', 'Sex', 'Delta 15 N (o/oo)', 'Delta 13 C (o/oo)', 'Comments'], axis= 1)
```

```python
df.style.set_caption("Aggregate Dataset").set_table_styles([{
    'selector': 'caption',
    'props': [
        ('color', 'navy'),
        ('font-size', '20px'),
        ("text-align", "center")
    ]
}])
```

```python
dfTable = df.agg(
    {
        "Culmen Length (mm)": ["mean","median", "std", "var", "min", "max","skew"],
           "Culmen Depth (mm)": ["mean","median", "std", "var", "min", "max","skew"],
          "Flipper Length (mm)": ["mean","median", "std", "var", "min", "max","skew"],
          "Body Mass (g)": ["mean","median", "std", "var", "min", "max","skew"],

    }
)
dfTable.style.set_caption("Aggregate Descriptive Stats").set_table_styles([{
    'selector': 'caption',
    'props': [
        ('color', 'navy'),
        ('font-size', '20px'),
        ("text-align", "center")
    ]
}])
```

```python
dfTable1 = df.iloc[:152].agg(
    {
        "Culmen Length (mm)": ["mean","median", "std", "var", "min", "max","skew"],
           "Culmen Depth (mm)": ["mean","median", "std", "var", "min", "max","skew"],
          "Flipper Length (mm)": ["mean","median", "std", "var", "min", "max","skew"],
          "Body Mass (g)": ["mean","median", "std", "var", "min", "max","skew"],

    }
)
dfTable2 = df.iloc[152:275].agg(
    {
        "Culmen Length (mm)": ["mean","median", "std", "var", "min", "max","skew"],
           "Culmen Depth (mm)": ["mean","median", "std", "var", "min", "max","skew"],
          "Flipper Length (mm)": ["mean","median", "std", "var", "min", "max","skew"],
          "Body Mass (g)": ["mean","median", "std", "var", "min", "max","skew"],

    }
)
dfTable3 = df.iloc[276:].agg(
    {
        "Culmen Length (mm)": ["mean","median", "std", "var", "min", "max","skew"],
           "Culmen Depth (mm)": ["mean","median", "std", "var", "min", "max","skew"],
          "Flipper Length (mm)": ["mean","median", "std", "var", "min", "max","skew"],
          "Body Mass (g)": ["mean","median", "std", "var", "min", "max","skew"],

    }
)
```
